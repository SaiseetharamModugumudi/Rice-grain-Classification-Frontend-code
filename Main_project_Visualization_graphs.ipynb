```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Define dataset path
dataset_path = r"C:\datasetfolder\rice_images"   

# Count images in each category
categories = ["Premium", "Regular", "Standard"]
image_counts = {category: len(os.listdir(os.path.join(dataset_path, category))) for category in categories}

# Plot Bar Chart
plt.figure(figsize=(8, 6))
sns.barplot(x=list(image_counts.keys()), y=list(image_counts.values()), palette="viridis")
plt.xlabel("Rice Grain Type")
plt.ylabel("Number of Images")
plt.title("Distribution of Rice Grain Types in Dataset")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(image_counts.values(), labels=image_counts.keys(), autopct="%1.1f%%", colors=["blue", "red", "green"])
plt.title("Percentage of Each Rice Grain Type")
plt.show()
```

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define dataset path
dataset_path =r"C:\datasetfolder\rice_images" 
categories = ["Premium", "Regular", "Standard"]

image_sizes = []

# Read image sizes using PIL instead of OpenCV
for category in categories:
    folder_path = os.path.join(dataset_path, category)
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        try:
            with Image.open(img_path) as img:
                width, height = img.size
                image_sizes.append(width * height)
        except Exception as e:
            print(f"Error loading image {img_name}: {e}")

# Plot histogram
plt.figure(figsize=(8, 6))
sns.histplot(image_sizes, bins=20, kde=True, color="blue")
plt.xlabel("Image Size (pixels)")
plt.ylabel("Count")
plt.title("Distribution of Image Sizes in Dataset")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Feature Extraction (Replace with actual extracted data)
data = {
    'Grain_Length': np.random.uniform(5.5, 7.5, 100),  # Example length values
    'Grain_Width': np.random.uniform(2.0, 3.5, 100),   # Example width values
    'Aspect_Ratio': np.random.uniform(1.5, 3.0, 100),  # Length-to-width ratio
    'Color_Intensity': np.random.uniform(50, 200, 100), # Color feature
    'Moisture_Content': np.random.uniform(10, 15, 100), # Moisture percentage
    'Defects_Presence': np.random.randint(0, 2, 100),   # 0 = No defect, 1 = Defective
    'Class': np.random.choice(['Full', 'Broken', 'Mixed'], 100) # Target labels
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Encode Categorical Target Variable (Full=0, Broken=1, Mixed=2)
df['Class'] = df['Class'].map({'Full': 0, 'Broken': 1, 'Mixed': 2})

# Compute Correlation Matrix
corr_matrix = df.corr()

# Plot the Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Rice Grain Attributes")
plt.show()
```
